```python
import pandas as pd

df = pd.read_excel('Areas-of-interest-POIs\\sample_1000.xlsx')
df.head()
```

```python
import geopandas as gpd

df1 = gpd.read_file('Areas-of-interest-POIs\\building_level_redistributed_values_with_fallback.gpkg')
df1.head()
```

```python
df["gml_id"] = df["gml_id"].astype(str)
df1["gml_id"] = df1["gml_id"].astype(str)

merged_df = df.merge(
    df1,
    on="gml_id",
    how="left",
    suffixes=("_sample", "_full")
)
```

```python
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df = merged_df[['mid_label_sample', 'mid_label_new', 'bosserhof_class_clean_sample',
       'bosserhof_class_clean.1', 'osm_names_sample', 'gml_id','volume_m3_sample', 'geometry']]
```

```python
merged_df.head()
```

```python
print('geometry' in merged_df.columns)
```

```python
merged_gdf = gpd.GeoDataFrame(
    merged_df,
    geometry='geometry',
    crs=df1.crs
)

merged_gdf.to_file(
    'Areas-of-interest-POIs\\validation_merged.gpkg',
    driver='GPKG'
)
```
