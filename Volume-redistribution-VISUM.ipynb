```python
import re
import pandas as pd
import numpy as np

valid_labels = [
    "Workers",
    "University",
    "School",
    "Kindergarten",
    "Retail_Daily",
    "Retail_Non-Daily",
    "Leisure",
]

label_map = {
    "work": "Workers",
    "business": "Workers",
    "university": "University",
    "school": "School",
    "childcare": "Kindergarten",
    "retail_daily": "Retail_Daily",
    "retail_non_daily": "Retail_Non-Daily",
    "errands": "Retail_Daily", # errands can be considered daily retail, but this is not true in all cases (public offices, banks, etc. need to decide about this)
    "leisure": "Leisure",
    "sports": "Leisure",
    "meetup": "Leisure",
    "lessons": "Leisure",
}

pattern = re.compile("|".join(sorted(map(re.escape, label_map.keys()), key=len, reverse=True)))


def clean_mid_label(x):
    if x is None:
        return []

    if isinstance(x, float) and pd.isna(x):
        return []

    if isinstance(x, (list, tuple, set, np.ndarray)):
        raw_labels = [str(v).strip() for v in x]
    else:
        raw_labels = pattern.findall(str(x))

    mapped = [label_map[label] for label in raw_labels if label in label_map]

    # remove duplicates, keep order
    return list(dict.fromkeys(mapped))


df["mid_label"] = df["mid_label"].apply(clean_mid_label)
```

```python
# ============================================================
# IMPORTS
# ============================================================

import ast
import numpy as np
import pandas as pd
import geopandas as gpd


# ============================================================
# CONFIG
# ============================================================

TAZ_ID_COL = "taz_uid"

CATEGORY_COLS = [
    "Workers",
    "School",
    "University",
    "Kindergarten",
    "Retail_Daily",
    "Retail_Non-Daily",
    "Leisure",
]

WORKER_COL = "Workers"

CATEGORY_TO_LABEL = {
    "Workers": "Workers",
    "School": "School",
    "University": "University",
    "Kindergarten": "Kindergarten",
    "Retail_Daily": "Retail_Daily",
    "Retail_Non-Daily": "Retail_Non-Daily",
    "Leisure": "Leisure",
}

# ============================================================
# WORKER CAP CONFIG
# ============================================================

STRICT_WORKER_CAP_CLASSES = [
    "restaurants gastronomy",
    "retail small scale",
    "fitness wellness",
    "entertainment culture",
    "large cinemas",
    "large discos fun leisure pools",
    "arenas large events",
    "theme parks",
    "car dealerships",
    "customer service",
    "vehicle electrical repair",
]

LARGE_FORMAT_RETAIL_CAP_CLASSES = [
    "shopping centers",
    "department stores",
    "self service department stores",
    "discount stores",
    "hypermarkets superstores",
    "diy stores",
    "furniture stores",
    "factory outlet centers",
]

# Change these values if needed
STRICT_CAP_QUANTILE = 0.75
LARGE_FORMAT_CAP_QUANTILE = 0.95

OUTPUT_GPKG = "Areas-of-interest-POIs/building_level_redistributed_values_with_fallback.gpkg"
VALIDATION_CSV = "Areas-of-interest-POIs/redistribution_validation_with_fallback.csv"
ALLOCATION_LOG_CSV = "Areas-of-interest-POIs/redistribution_allocation_log.csv"


# ============================================================
# HELPER: PARSE mid_label
# ============================================================

def parse_labels(x):
    if isinstance(x, list):
        return x

    if isinstance(x, np.ndarray):
        return list(x)

    if pd.isna(x):
        return []

    if isinstance(x, str):
        x = x.strip()

        if x in ["", "None", "nan", "NaN", "[]"]:
            return []

        try:
            parsed = ast.literal_eval(x)
            if isinstance(parsed, list):
                return parsed
        except Exception:
            pass

        x = x.strip("[]")
        return [
            v.strip().strip("'").strip('"')
            for v in x.split(",")
            if v.strip()
        ]

    return []


# ============================================================
# HELPER: CLEAN BOSSERHOF CLASS
# ============================================================

def clean_bosserhof_class(x):
    if pd.isna(x):
        return ""

    return (
        str(x)
        .lower()
        .strip()
        .replace("-", " ")
        .replace("_", " ")
        .replace("  ", " ")
    )


# ============================================================
# 1. PREPARE DATA
# ============================================================

taz = gdf.copy()
buildings = df.copy()

if not isinstance(taz, gpd.GeoDataFrame):
    taz = gpd.GeoDataFrame(taz, geometry="geometry")

if not isinstance(buildings, gpd.GeoDataFrame):
    buildings = gpd.GeoDataFrame(buildings, geometry="geometry", crs=taz.crs)

if buildings.crs != taz.crs:
    buildings = buildings.to_crs(taz.crs)

taz = taz[[TAZ_ID_COL, "geometry"] + CATEGORY_COLS].copy()
buildings = buildings.copy()

buildings["mid_label"] = buildings["mid_label"].apply(parse_labels)
buildings["volume_m3"] = pd.to_numeric(buildings["volume_m3"], errors="coerce").fillna(0)
buildings["potentials"] = pd.to_numeric(buildings["potentials"], errors="coerce").fillna(0)

buildings["bosserhof_class_clean"] = buildings["bosserhof_class"].apply(clean_bosserhof_class)

buildings = buildings.reset_index(drop=True)
buildings["building_index"] = buildings.index


# ============================================================
# 1A. CREATE DATA-DRIVEN WORKER VOLUME CAPS
# ============================================================

valid_cap_data = buildings[
    (buildings["volume_m3"] > 0) &
    (buildings["bosserhof_class_clean"] != "")
].copy()

STRICT_CLASS_VOLUME_CAPS = (
    valid_cap_data[
        valid_cap_data["bosserhof_class_clean"].isin(STRICT_WORKER_CAP_CLASSES)
    ]
    .groupby("bosserhof_class_clean")["volume_m3"]
    .quantile(STRICT_CAP_QUANTILE)
    .to_dict()
)

LARGE_FORMAT_CLASS_VOLUME_CAPS = (
    valid_cap_data[
        valid_cap_data["bosserhof_class_clean"].isin(LARGE_FORMAT_RETAIL_CAP_CLASSES)
    ]
    .groupby("bosserhof_class_clean")["volume_m3"]
    .quantile(LARGE_FORMAT_CAP_QUANTILE)
    .to_dict()
)

print("\nStrict worker caps:")
for k, v in STRICT_CLASS_VOLUME_CAPS.items():
    print(f"{k}: {v:.2f} m³")

print("\nLarge-format retail worker caps:")
for k, v in LARGE_FORMAT_CLASS_VOLUME_CAPS.items():
    print(f"{k}: {v:.2f} m³")


# ============================================================
# HELPER: WORKER EFFECTIVE WEIGHT
# ============================================================

def get_worker_effective_weight(row):
    """
    Only affects Workers.

    Strict classes:
        use class-specific 75th percentile volume cap

    Large-format retail classes:
        use class-specific 95th percentile volume cap

    Other classes:
        use original potentials unchanged
    """

    cls = row.get("bosserhof_class_clean", "")
    volume = row.get("volume_m3", 0)
    potentials = row.get("potentials", 0)

    if pd.isna(volume) or pd.isna(potentials):
        return 0

    if volume <= 0 or potentials <= 0:
        return 0

    if cls in STRICT_CLASS_VOLUME_CAPS:
        cap = STRICT_CLASS_VOLUME_CAPS[cls]
        effective_volume = min(volume, cap)
        return potentials * (effective_volume / volume)

    if cls in LARGE_FORMAT_CLASS_VOLUME_CAPS:
        cap = LARGE_FORMAT_CLASS_VOLUME_CAPS[cls]
        effective_volume = min(volume, cap)
        return potentials * (effective_volume / volume)

    return potentials


def get_worker_weight_method(row):
    cls = row.get("bosserhof_class_clean", "")

    if cls in STRICT_CLASS_VOLUME_CAPS:
        return "worker_strict_class_volume_capped_p75"

    if cls in LARGE_FORMAT_CLASS_VOLUME_CAPS:
        return "worker_large_format_volume_capped_p95"

    return "worker_potentials_original"


# ============================================================
# 2. ASSIGN BUILDINGS TO TAZ
# ============================================================

intersection = gpd.overlay(
    buildings,
    taz[[TAZ_ID_COL, "geometry"]],
    how="intersection"
)

intersection["intersect_area"] = intersection.geometry.area

intersection_sorted = intersection.sort_values(
    "intersect_area",
    ascending=False
)

best_match = intersection_sorted.drop_duplicates(
    subset=["building_index"]
)

buildings_taz = buildings.merge(
    best_match[["building_index", TAZ_ID_COL]],
    on="building_index",
    how="left"
)

buildings_taz = buildings_taz.rename(columns={TAZ_ID_COL: "target_taz"})


# ============================================================
# 3. CREATE BUILDING-ACTIVITY CANDIDATE TABLE
# ============================================================

candidate_rows = []

for _, row in buildings_taz.iterrows():
    labels = row["mid_label"]

    non_worker_labels = [
        lab for lab in labels
        if lab != CATEGORY_TO_LABEL[WORKER_COL]
    ]

    n_non_worker = len(non_worker_labels)

    for category, label in CATEGORY_TO_LABEL.items():

        if label not in labels:
            continue

        if category == WORKER_COL:
            weight = get_worker_effective_weight(row)
            weight_method = get_worker_weight_method(row)

        else:
            if n_non_worker == 0:
                continue

            weight = row["volume_m3"] / n_non_worker
            weight_method = "volume_split_by_non_worker_labels"

        if pd.isna(weight) or weight <= 0:
            continue

        candidate_rows.append({
            "building_index": row["building_index"],
            "gml_id": row.get("gml_id", None),
            "target_taz": row["target_taz"],
            "category": category,
            "weight": weight,
            "weight_method": weight_method,
            "volume_m3": row["volume_m3"],
            "potentials": row["potentials"],
            "bosserhof_class": row.get("bosserhof_class", None),
            "bosserhof_class_clean": row.get("bosserhof_class_clean", None),
        })

candidates = pd.DataFrame(candidate_rows)

print("\nCandidate activity rows:", len(candidates))

print("\nWorker candidate weight methods:")
print(
    candidates[candidates["category"] == WORKER_COL]["weight_method"]
    .value_counts(dropna=False)
)


# ============================================================
# 4. BUILD TAZ NEIGHBOR TABLE
# ============================================================

taz_for_neighbors = taz[[TAZ_ID_COL, "geometry"]].copy().reset_index(drop=True)

neighbors = gpd.sjoin(
    taz_for_neighbors,
    taz_for_neighbors,
    how="left",
    predicate="touches"
)

neighbors = neighbors[
    neighbors[f"{TAZ_ID_COL}_left"] != neighbors[f"{TAZ_ID_COL}_right"]
].copy()

neighbor_dict = (
    neighbors
    .groupby(f"{TAZ_ID_COL}_left")[f"{TAZ_ID_COL}_right"]
    .apply(list)
    .to_dict()
)


# ============================================================
# 5. REDISTRIBUTION WITH FALLBACK
# ============================================================

LOCAL_POOL_CATEGORIES = [
    "School",
    "University",
    "Kindergarten",
]

allocation_parts = []

for _, source_row in taz.iterrows():

    source_taz = source_row[TAZ_ID_COL]
    neighbor_tazs = neighbor_dict.get(source_taz, [])

    for category in CATEGORY_COLS:

        source_total = pd.to_numeric(source_row[category], errors="coerce")

        if pd.isna(source_total) or source_total == 0:
            continue

        if category in LOCAL_POOL_CATEGORIES:

            local_tazs = [source_taz] + neighbor_tazs

            selected = candidates[
                (candidates["category"] == category) &
                (candidates["target_taz"].isin(local_tazs))
            ].copy()

            assignment_level = "local_pool_same_plus_neighbor"

            if selected.empty:
                selected = candidates[
                    candidates["category"] == category
                ].copy()

                assignment_level = "global_fallback"

        else:

            selected = candidates[
                (candidates["category"] == category) &
                (candidates["target_taz"] == source_taz)
            ].copy()

            assignment_level = "within_taz"

            if selected.empty and len(neighbor_tazs) > 0:
                selected = candidates[
                    (candidates["category"] == category) &
                    (candidates["target_taz"].isin(neighbor_tazs))
                ].copy()

                assignment_level = "neighbor_taz"

            if selected.empty:
                selected = candidates[
                    candidates["category"] == category
                ].copy()

                assignment_level = "global_fallback"

        if selected.empty:
            allocation_parts.append(pd.DataFrame([{
                "source_taz": source_taz,
                "target_taz": None,
                "building_index": None,
                "gml_id": None,
                "category": category,
                "source_total": source_total,
                "weight": 0,
                "weight_sum": 0,
                "assigned_value": 0,
                "assignment_level": "unassigned_no_candidates",
                "weight_method": None,
            }]))
            continue

        weight_sum = selected["weight"].sum()

        if weight_sum <= 0:
            allocation_parts.append(pd.DataFrame([{
                "source_taz": source_taz,
                "target_taz": None,
                "building_index": None,
                "gml_id": None,
                "category": category,
                "source_total": source_total,
                "weight": 0,
                "weight_sum": 0,
                "assigned_value": 0,
                "assignment_level": "unassigned_zero_weight",
                "weight_method": None,
            }]))
            continue

        selected["source_taz"] = source_taz
        selected["source_total"] = source_total
        selected["weight_sum"] = weight_sum
        selected["assigned_value"] = selected["weight"] / weight_sum * source_total
        selected["assignment_level"] = assignment_level

        allocation_parts.append(selected)

allocation_log = pd.concat(allocation_parts, ignore_index=True)

allocation_log.to_csv(ALLOCATION_LOG_CSV, index=False)

print("\nAllocation rows:", len(allocation_log))
print(allocation_log["assignment_level"].value_counts(dropna=False))

print("\nWorker allocation weight methods:")
print(
    allocation_log[allocation_log["category"] == WORKER_COL]["weight_method"]
    .value_counts(dropna=False)
)


# ============================================================
# 6. AGGREGATE ASSIGNED VALUES BACK TO BUILDING LEVEL
# ============================================================

valid_allocations = allocation_log[
    allocation_log["building_index"].notna()
].copy()

valid_allocations["building_index"] = valid_allocations["building_index"].astype(int)

assigned_wide = valid_allocations.pivot_table(
    index="building_index",
    columns="category",
    values="assigned_value",
    aggfunc="sum",
    fill_value=0
)

assigned_wide.columns = [f"assigned_{c}" for c in assigned_wide.columns]

result = buildings_taz.merge(
    assigned_wide,
    left_on="building_index",
    right_index=True,
    how="left"
)

for c in CATEGORY_COLS:
    col = f"assigned_{c}"
    if col not in result.columns:
        result[col] = 0

    result[col] = result[col].fillna(0)


# ============================================================
# 7. VALIDATION BY SOURCE TAZ
# ============================================================

validation = (
    allocation_log
    .groupby(["source_taz", "category"], dropna=False)["assigned_value"]
    .sum()
    .reset_index()
    .rename(columns={"assigned_value": "assigned_sum"})
)

source_totals_long = taz[[TAZ_ID_COL] + CATEGORY_COLS].melt(
    id_vars=TAZ_ID_COL,
    value_vars=CATEGORY_COLS,
    var_name="category",
    value_name="source_total"
).rename(columns={TAZ_ID_COL: "source_taz"})

validation = validation.merge(
    source_totals_long,
    on=["source_taz", "category"],
    how="right"
)

validation["assigned_sum"] = validation["assigned_sum"].fillna(0)

validation["source_total"] = pd.to_numeric(
    validation["source_total"],
    errors="coerce"
).fillna(0)

validation["difference"] = (
    validation["assigned_sum"] -
    validation["source_total"]
)

validation.to_csv(VALIDATION_CSV, index=False)

print("\nMaximum absolute difference by source TAZ/category:")
print(validation["difference"].abs().max())

print("\nUnassigned rows:")
print(
    allocation_log[
        allocation_log["assignment_level"].isin([
            "unassigned_no_candidates",
            "unassigned_zero_weight"
        ])
    ][[
        "source_taz",
        "category",
        "source_total",
        "assignment_level"
    ]]
)


# ============================================================
# 8. SAVE FINAL BUILDING OUTPUT
# ============================================================

result = gpd.GeoDataFrame(
    result,
    geometry="geometry",
    crs=taz.crs
)

result.to_file(
    OUTPUT_GPKG,
    driver="GPKG"
)

print("\nSaved:")
print(OUTPUT_GPKG)
print(VALIDATION_CSV)
print(ALLOCATION_LOG_CSV)
```
